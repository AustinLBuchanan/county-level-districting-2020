```python
# imports
from gerrychain import Graph
from helper import add_2020_populations
from helper import check_feasibility
from helper import draw_districts
import math
import geopandas as gpd
```

```python
rows = list()

for state in congressional_seats_2020.keys():
        
    file = 'data\\' + state + '_county.json'
    
    # Read in the input graph (from 2010)
    G = Graph.from_json( file )
    k = congressional_seats_2020[state]   
    deviation = 0.01  # 1 percent
    
    if k > 1:
        # Compute the population lower and upper bounds under a +/- 0.5% deviation
        # To the graph object add the 2020 county populations 
        add_2020_populations( G, state )
        ideal = sum( G.nodes[node]['POP20'] for node in G.nodes ) / k
        L = math.ceil( ( 1 - deviation / 2 ) * ideal )
        U = math.floor( ( 1 + deviation / 2 ) * ideal )

        # Check if the instance is county-level feasible
        print("Testing if the counties of",state,"can be partitioned into",k,"contiguous districts, each with population between",L,"and",U)
        districts = check_feasibility(G, L, U, k)
        is_feasible = True if isinstance(districts, list) else False

        # Draw a feasible map if one exists.
        if is_feasible:
            file = 'data\\' + state + '_county.shp'
            df = gpd.read_file( file )
            draw_districts(G, districts, df)

        row = [ state, G.number_of_nodes(), k, L, U, is_feasible ]
    
    else:
        row = [ state, G.number_of_nodes(), k, 'ALL', 'ALL', 'Trivial']
        
    rows.append(row)
    print("")
```

```python
print("state n k L U feas?")
for row in rows:
    print(row)
```
